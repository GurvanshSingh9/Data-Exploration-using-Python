# **6. Bar Graph**


Bar graphs, like the one generated using Seaborn and Matplotlib in the provided code, visually display feature importances in a RandomForestClassifier.

 This aids in identifying influential features, enhancing interpretability, and guiding feature selection for model optimization.



```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Assuming X is your feature matrix, y is your target variable, and df is your DataFrame
X = df.drop('class', axis=1)
y = df['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances.index, y=feature_importances.values, palette='viridis')
plt.title('Feature Importances')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.xticks(rotation=45, ha='right')
plt.show()

```

# **7. Overlaid Histograms**

The provided code utilizes Seaborn to create Overlaid Histograms, displaying the distribution of numeric features in a DataFrame.

This visualization method combines multiple histograms, offering insights into the data's variable-specific frequency distributions in a single plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns

# Use Seaborn to create histograms
plt.figure(figsize=(12, 10))
for feature in numeric_features:
    sns.histplot(df[feature], kde=True, bins=30, label=feature)

plt.title('Histograms of Numeric Features')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

```

# **11. pairplot of selected features:**


A Pairplot visually explores relationships among selected features, with class differentiation.

This code generates a pairplot for pregnancies, glucose concentration, diastolic blood pressure, and class, aiding in comprehensive data analysis and pattern recognition.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
selected_features = ['Pregnancies', 'PG_Concentration', 'Diastolic_BP', 'class']

# Create a pairplot
sns.pairplot(df[selected_features], hue='class', markers=['o', 's'], palette='viridis')
plt.suptitle('Pairplot of Selected Features', y=1.02, ha='center', fontsize='large')
plt.show()
```
